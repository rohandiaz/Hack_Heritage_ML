```python
# Define required skills
required_skills = [
    "Python", "SQL", "Machine Learning", "Data Science"
]

def extract_text_from_pdf(pdf_file):
    """Extract text from a PDF file."""
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text() or ""
    return text

def preprocess_text(text):
    """Preprocess text by lowering the case and removing non-alphanumeric characters."""
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    return text

def extract_skills(text, skills):
    """Extract and match skills from the text using contextual analysis and pattern matching."""
    matched_skills = []
    text = preprocess_text(text)

    for skill in skills:
        skill_lower = skill.lower()
        if re.search(r'\b' + re.escape(skill_lower) + r'\b', text):
            matched_skills.append(skill)

    return matched_skills

def infer_skills(matched_skills, job_requirements):
    """Infer additional skills based on matched skills and job requirements."""
    inferred_skills = set(matched_skills)
    job_skills = set(job_requirements.split(', '))

    for skill in matched_skills:
        for category, related_skills in skill_inference.items():
            # Only infer skills that are relevant to the job requirements
            relevant_skills = job_skills.intersection(related_skills)
            if relevant_skills:
                inferred_skills.update(relevant_skills)

    return ', '.join(inferred_skills)



def process_resumes_from_directory(directory_path, job_requirements):
    """Process all resumes in a directory and rank them based on job requirements."""
    resumes = []

    for filename in os.listdir(directory_path):
        if filename.lower().endswith(".pdf"):
            file_path = os.path.join(directory_path, filename)
            with open(file_path, "rb") as file:
                text = extract_text_from_pdf(file)
                matched_skills = extract_skills(text, required_skills)
                inferred_skills = infer_skills(matched_skills, job_requirements)

                # Combine matched and inferred skills
                all_skills = list(set(matched_skills + inferred_skills.split(', ')))
                resumes.append({
                    "filename": filename,
                    "text": text,
                    "matched_skills": ', '.join(matched_skills),
                    "inferred_skills": inferred_skills,
                    "all_skills": ', '.join(all_skills)
                })

    if not resumes:
        print("No resumes were processed. Check the directory path or PDF files.")
        return

    # Combine job requirements with inferred skills
    all_skills = [resume['all_skills'] for resume in resumes] + [job_requirements]

    if not all_skills:
        print("No skills extracted. Check the resume parsing and skill extraction.")
        return

    # Vectorize the skills
    vectorizer = CountVectorizer(tokenizer=lambda x: x.split(', '))
    skill_vectors = vectorizer.fit_transform(all_skills)

    if skill_vectors.shape[0] <= 1:
        print("Insufficient data for cosine similarity computation.")
        return

    # Compute cosine similarity
    cosine_sim = cosine_similarity(skill_vectors[-1], skill_vectors[:-1])

    if cosine_sim.size == 0:
        print("Cosine similarity computation resulted in an empty array.")
        return

    # Add similarity scores to the resumes
    for i, resume in enumerate(resumes):
        resume['match_score'] = cosine_sim.flatten()[i]

    # Rank resumes by match score
    sorted_resumes = sorted(resumes, key=lambda x: x['match_score'], reverse=True)

    # Output ranked resumes
    df = pd.DataFrame(sorted_resumes)
    print("\nRanked Resumes:")
    print(df[['filename', 'inferred_skills', 'match_score']])

    # Plotting the pie chart
    plt.figure(figsize=(10, 7))
    plt.pie(df['match_score'], labels=df['filename'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(df))))
    plt.title('Proportion of Match Score for Each Resume')
    plt.show()

# Example usage
if __name__ == "__main__":
    directory_path = "/content"  # Adjust path to where your resumes are located
    job_requirements = "Python, SQL, Machine Learning, Data Science"
    process_resumes_from_directory(directory_path, job_requirements)

```
